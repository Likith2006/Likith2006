```python
#word tokenize
from nltk.tokenize import word_tokenize
sentence = ("To containerize an ML model means to package the model and all its dependencies, inculding libries and runtime environments, into a container is a lightweight and portable package that contains everything needed to run an application or service.")

words = word_tokenize(sentence)
print(words)
```

```python
#Visualize the Frequency Distribution of Words
import nltk
import matplotlib.pyplot as plt
from nltk.probability import FreqDist

text = ("To containerize an ML model means to package the model and all its dependencies, inculding libries and runtime environments, into a container is a lightweight and portable package that contains everything needed to run an application or service.")

words = word_tokenize(text)
freq_dist = FreqDist(words)
freq_dist.plot(20)
plt.show()
```

```python
#droping stop word
import spacy

nlp = spacy.load("en_core_web_sm")
text = ("To containerize an ML model means to package the model and all its dependencies, inculding libries and runtime environments, into a container is a lightweight and portable package that contains everything needed to run an application or service.")

doc = nlp(text)
filtered_words = [token.text for token in doc if not token.is_stop]
```

```python
#droping punctuation
import string
text = ("To containerize an ML model means to package the model and all its dependencies, inculding libries and runtime environments, into a container is a lightweight and portable package that contains everything needed to run an application or service.")

cleaned_text = text.translate(str.maketrans('', '', string.punctuation))
```

```python
#stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["containerize", "means", "package"]
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

```python
#lemmatization
import spacy
nlp = spacy.load("en_core_web_sm")
words = ["containerize", "means", "package"]
lemmatized_words = [token.lemma_ for token in nlp(" ".join(words))]
print(lemmatized_words)
```

```python
#bigrams
from nltk.util import ngrams
from nltk import word_tokenize
text =("To containerize an ML model means to package the model and all its dependencies, inculding libries and runtime environments, into a container is a lightweight and portable package that contains everything needed to run an application or service.")

tokens = word_tokenize(text)
bigrams = list(ngrams(tokens, 2))
print(bigrams)
```

```python
#trigrams
trigrams = list(ngrams(tokens, 3))
print(trigrams)
```

```python
#unigrams
unigrams = list(ngrams(tokens, 1))
print(unigrams)
```
